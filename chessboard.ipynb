```python
%matplotlib inline
import  matplotlib.pyplot as plt
```

```python
class Board:
    def __init__(self):
        self.grid = np.zeros((8,8,3))
        self.grid[1::2, ::2] = 1
        self.grid[::2, 1::2] = 1
        self.blue_row = None
        self.blue_column = None
        self.red_row = None
        self.red_column = None

    def add_red(self, row, col):
        self.red_row = row
        self.red_column = col
        self.grid[row][col] = (1,0,0)

    def add_blue(self, row, col):
        self.blue_row = row
        self.blue_column = col
        self.grid[row][col] = (0,0,1)
        
    def clear(self):
        self.grid = np.zeros((8,8,3))
        self.grid[1::2, ::2] = 1
        self.grid[::2, 1::2] = 1
        self.blue_row = None
        self.red_column = None
        self.red_row = None
        self.blue_column = None
        
    def render(self):
        plt.imshow(self.grid)
        

    def is_under_attack(self):
        if self.blue_row == self.red_row or self.blue_column == self.red_column:
            return True

        if abs(self.red_column - self.blue_column) == abs(self.red_row - self.blue_row):
            return True
        
        return False
```

```python


chess_board = Board()
chess_board.add_red(2,2)
chess_board.add_blue(7,4)
chess_board.render()
print(chess_board.is_under_attack())


     
```

# Queens on same row show "under attack"

```python
chess_board.clear()
chess_board.add_red(6,2)
chess_board.add_blue(6,4)
chess_board.render()
assert chess_board.is_under_attack()
print('UNDER ATTACK')
```

# Queens on same column show "under attack"

```python
chess_board.clear()
chess_board.add_red(1,4)
chess_board.add_blue(6,4)
chess_board.render()
assert chess_board.is_under_attack()
print('UNDER ATTACK')
```

# Queens on same diagonal should be “under attack”

```python
chess_board.clear()
chess_board.add_red(1,4)
chess_board.add_blue(4,1)
chess_board.render()
assert chess_board.is_under_attack()
print('UNDER ATTACK')
```

# Queens with any other coordinates should NOT be “under attack"

```python
chess_board.clear()
chess_board.add_red(2,5)
chess_board.add_blue(3,7)
chess_board.render()
assert not chess_board.is_under_attack()
print('NOT UNDER ATTACK')
     
```
